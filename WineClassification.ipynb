The wine dataset is the result of a chemical analysis of wines produced in the same region in Italy but derived from three different cultivars. (For illustration simplicity purpose, only 2 classes, 0 and 1, will be included for the classification task.) The analysis determined the quantities of 13 constituents found in each of the three types of wines.

The objective is to classify the wines into class 0 or 1 using the 13 given attributes and k-NN classifier.

```python
from sklearn import datasets
import pandas as pd
import numpy as np
from numpy import arange
import seaborn as sns
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
import sklearn
import math
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# load the wine dataset 
wine = datasets.load_wine()
print(wine.DESCR)
```

```python
# convert the data into dataframe format
X = pd.DataFrame(wine['data'], columns = wine['feature_names'])
y = wine['target']

# only consider wine class 0 and 1
X = X.loc[0:129, :]
y = y[0:130]

X.head()      
```

```python
data = X
data['target'] = y
data
```

```python
#Standardizing and splitting into train and test 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test= train_test_split(X.iloc[:,:-1],y, test_size=0.25, random_state= 200)
scaler= preprocessing.StandardScaler()
X_train_s=pd.DataFrame(scaler.fit_transform(X_train), columns= X.iloc[:,:-1].columns)
X_test_s=pd.DataFrame(scaler.fit_transform(X_test), columns = X.iloc[:, :-1].columns)
print('X:', len(X), 'X_train :', len(X_train), 'X_test :', len(X_test))
```

```python
#Determining the best k value based on classification accuracy on validation set
results = []
for k in range(1, 10):
  knn = KNeighborsClassifier(n_neighbors=k).fit(X_train_s, y_train)
  results.append({
  'k': k,
  'accuracy': accuracy_score(y_test, knn.predict(X_test_s))
  })
# Convert results to a pandas data frame
results = pd.DataFrame(results)
print(results)
#The best k is k=7 
```

```python
#Classifying new record for the chosen k
# New record
new_wine = pd.DataFrame(columns = wine['feature_names'])
new_wine.loc[0,:] = np.array([14.12, 1.88, 2.31, 18.5, 125, 2.50, 3.12, 0.26, 2.12, 4.87, 1.02, 3.23, 955])
new_wine
```

```python
#  since the data has only 130 records it is not necessary to perform data partition 
#  before scoring the new record, to prevent overfitting to training data set. If the number of records were higher, we might 
# need to reduce the records to reduce computation complexity
X= X.drop('target', axis=1)
new_wine_scale= new_wine-X.mean()/X.std()
knn = KNeighborsClassifier(n_neighbors=7).fit(X,y)
new_wine_pred= knn.predict(new_wine_scale)
new_wine_pred
```
